# Define and plot two vectors

Here is the code to plot vectors `v1` and `v2` as x and y coordinate vectors, and name the plot "title":
```
plot(x=v1, y=v2, main="title")
```
Your turn!

- Define two sample vectors `foo` and `bar`.
- The elements of `foo` are: `1.1`, `2`, `3.5`, `3.9`, `4.2`.
- The elements of `bar` are: `2`, `2.2`, `-1.3`, `0`, `0.2`.
- Plot these vectors with `plot()`, so that `foo` is on the x-axis, and `bar` on the y-axis
- Call the plot "Vectorplot"

````r
## define two vectors and plot them
foo  <- c(1.1, 2, 3.5, 3.9, 4.2)
bar <- c(2,2.2,-1.3,0,0.2)
plot(x=foo, y=bar, main="Vectorplot") 
````

# Graphical parameters

`plot()` has many arguments. They are passed to the function inside the parentheses, separated by commata:
```
plot(x,y,arg1=[...],arg2=[...],...)
```
For a complete list, check the help or the [cheat sheet](http://publish.illinois.edu/johnrgallagher/files/2015/10/BaseGraphicsCheatsheet.pdf) (PDF).


## Line and point `type`

Set a `type` attribute: 
```
plot(x,y,type=[attribute])
```
Available attribute values:
   * `"l"` for line plots
   * `"b"` for both lines and (data) points
   * `"o"` for plotting points over lines
   * `"n"` for no points whatsoever

### Lines only

Plot `foo` and `bar` as a line plot.

````r
plot(foo,bar,type="l")
````

### Lines and points
Plot `foo` and `bar` with both points and lines.

````r
plot(foo,bar,type="b")
````

### Lines plotted over points

Plot `foo` and `bar` as a plot with lines plotted over the points.

````r
plot(foo,bar,type="o")
````

### No points

Sometimes you want to build up a plot in stages (we'll do this later). Then it's useful to create a plot with no points - actually the points are still there (as you can see from the axes), but they're not visible.

Plot `foo` and `bar` as a plot without points, or lines.

````r
plot(foo,bar,type="n")
````

## Title and axis labels `main`, `xlab` and `ylab`

- Plot title: `main="title"`
- Plot x-axis label: `xlab="x label"`
- Plot y-axis label: `ylab="y label"`

To define a title or a label over more than one line, use the escape character `\n`. 



Plot the vectors `foo` and `bar` as a line plot with points, with the title `"My lovely plot"`, and the x- and y-axis labels `foo` and `bar`, respectively.

````r
plot(foo,bar,
main="My lovely plot",
xlab="foo",
ylab="bar")
````

Plot the vectors again in the same way, but this time, print `My lovely plot` in the first, and `over two lines` in the second line of the title. 

````r
plot(foo,bar,
main="My lovely plot\nover two lines",
xlab="foo",
ylab="bar")
````

## Line and point appearances: `pch`, `lty`, `lwd`, `cex`


### Change point characters with `pch`
- The `pch` attribute determines point appearance
- `pch` assumes values between `1` and `25`
- Get more information with `?points`


To show all available point appearances, store the vector `1:25` as both x and y argument of `plot()`, title the plot `pch demo`, and set the `pch` attribute to `1:25'. 

This will draw the available point characters on a diagonal line.

````r
plot(1:25,1:25,main="pch demo",pch=1:25)
````

To get rid of the x- and y-axis labels, define `xlab=""` and `ylab=""` (empty string).

### Change line type with `lty`
- The `lty` attribute determines the line type
- You can assign the values `1` through `6` to `lty`


First, plot the vector `1:6` (as the only data argument), set the title to `lty demo`, remove the x- and y-axis labels, and draw no points (so the plot itself appears empty).

````r
plot(1:6,main="lty demo", xlab="",ylab="",type="n")
````

Next, we use the function `abline()`. It draws vertical or horizontal straight lines through a plot. We're going to use it to demonstrate the line types. Enter the following command:
```
abline(v=(=1:6,
       col=c("blue","red","seagreen", 
             "black","orange","aquamarine4"),
       lty=1:6)
``` 
- The first argument specifies the x-values for the vertical lines. 
- The 2nd argument colors each line in a different color. 
- The third argument aligns each drawn vertical line with a different line type 

````r
plot(1:6,main="lty demo", xlab="",ylab="",type="n") + 
abline(v=1:6,
       col=c("blue","red","seagreen", 
             "black","orange","aquamarine4"),
       lty=1:6)
````

*Note: in the online notebook, you have to enter the previous `plot()` command again immediately before the `abline()` command and connect the two with a `+`:* 
```
plot(...) + abline(...)
```
*This has to do with the graphics device control of R (you can list currently open devices with `dev.list()`).*

````r
dev.list()
````

### Change point extension with `cex`
- The `cex` attribute sets the plotted point size.
- The value of `cex` must be positive.

We proceed as before, using a few of the arguments we have already used.



Plot the vector `1:8` as the only argument (i.e. a diagonal of 8 points), and
- set the title to `cex=2, pch=16`
- remove x- and y-axis labels
- color each point differently using `1:8` for the color attribute
- choose the point character `16`
- choose the point extension `2`

````r
plot(1:8,
main="cex=2, pch=16",
xlab="",
ylab="",
col=1:8,
pch=16,
cex=2)
````

### Change line width with `lwd`
- Line thickness is set with the `lwd` attribute
- the value of `lwd` must be positive

Re-use the code from the `lty` demonstration earlier, and add inside `abline()` the argument `lwd=4`. The output is the same, with thicker lines.


````r
plot(1:6,main="lty demo", xlab="",ylab="",type="n") + 
abline(v=1:6,
       col=c("blue","red","seagreen", 
             "black","orange","aquamarine4"),
       lty=1:6,
       lwd=4)
````

## Plotting region limits

- R chooses displayed plot ranges according to the input values
- Custom plot ranges can be set using the arguments `xlim` and `ylim`
- Their values are set as vectors with `c()`

Plot `foo` and `bar` with the custom ranges:
```
xlim=c(-10,5)
ylim=c(-3,3)
```
Use the following attributes:
- title: `custom area demo`
- axis labels: empty
- point color: `blue`
- point character: `8`
- line type: `2`
- point extension: `2.3`
- line width: `3.3`

````r
 plot(foo, bar, type="b",
          main="custom area\ndemo",
          xlab="",
          ylab="",
          col="blue",
          pch=8,
          lty=2,
          cex=2.3,
          lwd=3.3,
          xlim=c(-10,5), # x axis
          ylim=c(-3,3))  # y axis
````

You are not tied to the input values `x` and `y`. Re-use the previous plot, but with a different custom range:
``` 
xlim=c(3,5)
ylim=c(-0.5,0.2)
```

````r
plot(foo, bar, type="b",
          main="custom area\ndemo",
          xlab="",
          ylab="",
          col="blue",
          pch=8,
          lty=2,
          cex=2.3,
          lwd=3.3,
          xlim=c(3,5), # x axis
          ylim=c(-0.5,0.2))  # y axis
````

You see that some of the values cannot be displayed at all within the given custom range.
